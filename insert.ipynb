```python
import pandas as pd
import mysql.connector
import pymysql
```

```python

data = pd.read_csv('used_bikes.csv')  
data['price'] = data['price'].astype(float)
data['kms_driven'] = data['kms_driven'].astype(int)
```

```python
data.head()
```

```python
memory_before = data.memory_usage(deep=True).sum() / (1024 * 1024) 
```

```python
db_connection = mysql.connector.connect(
    host='127.0.0.1',
    port = '3306',
    user='root',
    password='1234',
    database='task2'
)

cursor = db_connection.cursor()
```

```python






# Step 4: Create a table in the database
create_table_query = """
CREATE TABLE IF NOT EXISTS new_bikes (
    id INT AUTO_INCREMENT PRIMARY KEY,
    bike_name VARCHAR(255),
    price VARCHAR(255),
    city VARCHAR(255),
    kms_driven VARCHAR(255),
    owner VARCHAR(255),
    age VARCHAR(255),
    power VARCHAR(255),
    brand VARCHAR(255)
)
"""

cursor.execute(create_table_query)

# Step 5: Insert values into the table
for _, row in data.iterrows():
    insert_query = """
    INSERT INTO new_bikes (bike_name, price, city, kms_driven, owner, age, power, brand)
    VALUES (%s, %s, %s, %s, %s, %s, %s, %s)
    """
    values = (
        row['bike_name'],
        row['price'],
        row['city'],
        row['kms_driven'],
        row['owner'],
        row['age'],
        row['power'],
        row['brand']
    )
    cursor.execute(insert_query, values)

# Step 6: Index the 'bike_name' column (you can replace it with your desired column)
index_query = "CREATE INDEX bike_name_index ON new_bikes (bike_name)"
cursor.execute(index_query)

# Commit changes and close connections
db_connection.commit()
cursor.close()
db_connection.close()

# Check memory usage after changing data types
memory_after = data.memory_usage(deep=True).sum() / (1024 * 1024)  # Memory in MB

print(f"Memory usage before changing data types: {memory_before:.2f} MB")
print(f"Memory usage after changing data types: {memory_after:.2f} MB")
```
